```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings 
warnings.filterwarnings("ignore")
```

```python
df=sns.load_dataset("iris")
```

```python
df.head()
```

```python
df["species"].unique()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=df.drop("species",axis=1)
y=df["species"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model_knn=KNeighborsClassifier(n_neighbors=13)
```

```python
model_knn.fit(X_train,y_train)
```

```python
model_knn.score(X_test,y_test)
```

```python
from sklearn.svm import SVC
```

```python
model_svm=SVC(C= 10,kernel="linear",gamma="auto")
```

```python
model_svm.fit(X_train,y_train)
```

```python
model_svm.score( X_test,y_test)
```

```python
#lets use the grid search cv with svm model
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
classifier= GridSearchCV((model_svm),{
    "C":[1,10,20,30],
    "kernel":["rbf","linear"],
  },cv=5,return_train_score=False)
```

```python
classifier.fit(x,y)
```

```python
print(classifier.cv_results_)
```

```python
results=pd.DataFrame(classifier.cv_results_)
```

```python
results
```

```python
results[["param_C","param_kernel","mean_test_score"]]
```

```python
# now we use the grid search cv with KNN  model
```

```python
classifier_2 = GridSearchCV((model_knn),{
    "n_neighbors":[3,5,9,13],
    "weights":["uniform", "distance"],
    "algorithm":["auto","ball_tree","kd_tree","brute"]
},cv=5,return_train_score=False)
```

```python
classifier_2.fit(x,y)
```

```python
print(classifier_2.cv_results_)
```

```python
results=pd.DataFrame(classifier_2.cv_results_)
```
